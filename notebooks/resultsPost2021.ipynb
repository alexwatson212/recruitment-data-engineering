```python
import pandas as pd
import numpy as np
from io import BytesIO
from minio import Minio
from sqlalchemy import create_engine, text
```

```python
# Initialize Minio client
minio_client = Minio(
    "minio:9000",
    access_key="minioadmin",
    secret_key="minioadmin",
    secure=False
)

# Download the CSV files from the Minio bucket into a pandas DataFrame
data = minio_client.get_object("track.data-raw", "results.csv")
data = BytesIO(data.read())
result_csv = pd.read_csv(data)

#Due to Postgres foreign key restraints other csv files must be loaded into repsective tables before results may be loaded in.
data1 = minio_client.get_object("track.data-raw", "races.csv")
data1 = BytesIO(data1.read())
race_csv = pd.read_csv(data1)

data2 = minio_client.get_object("track.data-raw", "drivers.csv")
data2 = BytesIO(data2.read())
drivers_csv = pd.read_csv(data2)

data3 = minio_client.get_object("track.data-raw", "constructors.csv")
data3 = BytesIO(data3.read())
constructors_csv = pd.read_csv(data3)

data4 = minio_client.get_object("track.data-raw", "status.csv")
data4 = BytesIO(data4.read())
status_csv = pd.read_csv(data4)
```

```python
#Insert Status records into Postgres 
```

```python
for index, row in status_csv.iterrows():
    csv_status = row[['statusId', 'status']]
    insert_query = text("""
        INSERT INTO race_data.status 
        (status_id, status) 
        VALUES (:statusId, :status)
        """)
    csv_status_dict = csv_status.to_dict()
    with engine.begin() as connection:
        connection.execute(insert_query, csv_status_dict)
```

```python
#Insert Constructor records into Postgres 
```

```python
for index, row in constructors_csv.iterrows():
    csv_constructors = row[['constructorId', 'name']]
    insert_query = text("""
        INSERT INTO race_data.constructors 
        (constructor_id, name) 
        VALUES (:constructorId, :name)
        """)
    csv_constructors_dict = csv_constructors.to_dict()
    with engine.begin() as connection:
        connection.execute(insert_query, csv_constructors_dict)
```

```python
#Insert Drivers records into Postgres 
```

```python
for index, row in drivers_csv.iterrows():
    csv_drivers = row[['driverId', 'driverRef', 'code', 'forename', 'surname', 'dob']]
    insert_query = text("""
        INSERT INTO race_data.drivers 
        (driver_id, driver_ref, code, forename, surname, dob) 
        VALUES (:driverId, :driverRef, :code, :forename, :surname, :dob)
        """)
    csv_drivers_dict = csv_drivers.to_dict()
    with engine.begin() as connection:
        connection.execute(insert_query, csv_drivers_dict)
```

```python
#Handle \N values in position 
```

```python
results2021['position'] = results2021['position'].replace(to_replace = '\\N', value = 99999) 
```

```python
#Handle \N values in time column and replace with +99:99.999
```

```python
results2021['time'] = results2021['time'].replace(to_replace = '\\N', value = '+99:99.999') 
```

```python
#Handle \N values in fastestLapTime column and replace with +99:99.999
```

```python
results2021['fastestLapTime'] = results2021['fastestLapTime'].replace(to_replace = '\\N', value = '+99:99.999') 
```

```python
#Handle \N values in fastestLap column and replace with -1
```

```python
results2021['fastestLap'] = results2021['fastestLap'].replace(to_replace = '\\N', value = '-1') 
```

```python
#Handle \N values in fastestLapSpeed column and replace with 000.000
```

```python
results2021['fastestLapSpeed'] = results2021['fastestLapSpeed'].replace(to_replace = '\\N', value = '999.999') 
```

```python
#Handle \N values in milliseconds column and replace with -1
```

```python
results2021['milliseconds'] = results2021['milliseconds'].replace(to_replace = '\\N', value = '-1' ) 
```
